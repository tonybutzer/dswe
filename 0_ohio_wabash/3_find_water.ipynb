```python
import numpy as np
import tqdm.notebook as tq
import zarr
```

```python
H='21'
V='10'
```

```python
Y=0
X=0
Z_store = f's3://eccoe-scratch/dwse/{H}{V}_{Y}{X}.zarr_store'
```

```python
# To load the Zarr array back into a NumPy array
loaded_array = zarr.load(Z_store)

# Verify the loaded array
print(loaded_array.shape)
```

```python
def count_ones(ary):
    count_of_ones = np.count_nonzero(ary == 1)
    return count_of_ones
```

```python
nlen = loaded_array.shape[0]
```

```python
%%time
for t1 in range(nlen):
    cnt_ones = count_ones(loaded_array[t1])
    print(cnt_ones)
```
